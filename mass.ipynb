## MASS Scale's Survey
MASS Scale means psychometric of detecting Machiavellianism score.  
It have 20 questions and 5 answers.  

questions.  
"Q1" : "Never tell anyone the real reason you did something unless it is useful to do so.",  
"Q2" : "The best way to handle people is to tell them what they want to hear.",  
"Q3" : "One should take action only when sure it is morally right.",  
"Q4" : "Most people are basically good and kind.",  
"Q5" : "It is safest to assume that all people have a vicious streak and it will come out when they are given a chance.",  
"Q6" : "Honesty is the best policy in all cases.",  
"Q7" : "There is no excuse for lying to someone else.",  
"Q8" : "Generally speaking, people won't work hard unless they're forced to do so.",  
"Q9" : "All in all, it is better to be humble and honest than to be important and dishonest.",  
"Q10" : "When you ask someone to do something for you, it is best to give the real reasons for wanting it rather than giving reasons which carry more weight.",  
"Q11" : "Most people who get ahead in the world lead clean, moral lives.",  
"Q12" : "Anyone who completely trusts anyone else is asking for trouble.",  
"Q13" : "The biggest difference between most criminals and other people is that the criminals are stupid enough to get caught.",  
"Q14" : "Most people are brave.",  
"Q15" : "It is wise to flatter important people.",  
"Q16" : "It is possible to be good in all respects.",  
"Q17" : "P.T. Barnum was wrong when he said that there's a sucker born every minute.",  
"Q18" : "It is hard to get ahead without cutting corners here and there.",  
"Q19" : "People suffering from incurable diseases should have the choice of being put painlessly to death.",  
"Q20" : "Most people forget more easily the death of their parents than the loss of their property."  

answers.(have five point scale)  
1=Sisagree, 2=Slightly disagree, 3=Neutral, 4=Slightly agree, 5=Agree.  

calcute total scale.  
sum question's answer.  

this question are composed of 2 facet.  
1. true question: answers point means Machiavellianism score.  
(for example. Sisagree means 1 point. Agree means 5 point)  
2. reverse question: (6 - answers point) means Machiavellianism score.  
(for example. Sisagree means 5 point. Agree means 1 point)  

reverse question: ['Q6A', 'Q7A', 'Q14A', 'Q4A', 'Q17A', 'Q16A', 'Q9A', 'Q10A', 'Q11A', 'Q3A']  
true question: ['Q1A', 'Q2A', 'Q5A', 'Q8A', 'Q12A', 'Q13A', 'Q15A', 'Q18A', 'Q19A', 'Q20A']  

Detail of Machiavellianism.  
https://en.wikipedia.org/wiki/Machiavellianism_(psychology)  

```python
import pandas as pd

df = pd.read_table('MACH_data\data.csv')
df
```

```python
# extract high school students.

pd.set_option('display.max_columns', 500)
df.head()
df_aft = df[df['education'] == 2]
df_aft = df_aft[df_aft['age'] < 20]
df_aft
```

```python
# set true question, reverse question
indexes = list(range(1, 20+1))
reverse_index = [6, 7, 14, 4, 17, 16, 9, 10, 11, 3]
true_index = [i for i in indexes if i not in reverse_index]
reverse_index_str = [f'Q{i}A' for i in reverse_index]
true_index_str = [f'Q{i}A' for i in true_index]

reverse_index_str, true_index_str
```

```python
pd.set_option('display.max_rows', 500)
```

```python
from matplotlib import pyplot as plt

plt.rcParams["figure.figsize"] = (20, 6)
```

## MASS Scale per country.

```python
df_country = df.groupby('country').mean()
df_result = df_country[df.groupby('country').size() > 100]
scale_series = df_result[true_index_str].sum(axis=1) - df_result[reverse_index_str].sum(axis=1)

scale_series.plot.bar()
```

## MASS Scale per country by high school student.

```python
df_country = df_aft.groupby('country').mean()
df_result = df_country[df_aft.groupby('country').size() > 40]
scale_series = df_result[true_index_str].sum(axis=1) - df_result[reverse_index_str].sum(axis=1)

scale_series.plot.bar()
```

# gender: "What is your gender?", 1=Male, 2=Female, 3=Other

```python
df_gb = df_aft.groupby(['country', 'gender'])
df_country_gender = df_gb.mean()
df_result = df_country_gender[df_gb.size() > 20]
scale_series = df_result[true_index_str].sum(axis=1) - df_result[reverse_index_str].sum(axis=1)

scale_series.plot.bar()
```

## MASS scale by ages



```python
df_gb = df_aft.groupby(['country', 'age'])
df_country_age = df_gb.mean()
df_result = df_country_age[df_gb.size() > 50]
scale_series = df_result[true_index_str].sum(axis=1) - df_result[reverse_index_str].sum(axis=1)

scale_series.plot.bar()
```
